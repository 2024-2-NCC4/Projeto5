```python
# Função para converter valores numéricos com vírgulas para pontos decimais
def converter_valores(valor):
    if isinstance(valor, str):
        valor = valor.replace('.', '').replace(',', '.')
        if 'K' in valor:
            return float(valor.replace('K', '')) * 1_000
        elif 'M' in valor:
            return float(valor.replace('M', '')) * 1_000_000
        elif 'B' in valor:
            return float(valor.replace('B', '')) * 1_000_000_000
        else:
            return float(valor)
    return valor
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do CSV
df = pd.read_csv('historico_AAPL.csv')

# Converter a coluna 'Data' para o tipo datetime, especificando que o dia vem primeiro
df['Data'] = pd.to_datetime(df['Data'], dayfirst=True)

# Aplicar a conversão de valores nas colunas numéricas
df['Último'] = df['Último'].apply(converter_valores)
df['Abertura'] = df['Abertura'].apply(converter_valores)
df['Máxima'] = df['Máxima'].apply(converter_valores)
df['Mínima'] = df['Mínima'].apply(converter_valores)
df['Vol.'] = df['Vol.'].apply(converter_valores)

# Definir a 'Data' como índice para facilitar a plotagem
df.set_index('Data', inplace=True)

# Plotar o gráfico com o desempenho (Último)
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Último'], label='Último', linestyle='-', color='black')

# Personalizações do gráfico
plt.title('Desempenho da Empresa na Bolsa')
plt.xlabel('Data')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)

# Exibir o gráfico
plt.show()
```
